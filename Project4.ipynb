## Loading Data

We will fetch the first two chapters of Jane Austen's Pride and Prejudice from [Project Gutenberg](https://www.gutenberg.org/ebooks/1342)


### Function to fetch data


```python
def fetch_text(raw_url):
  """
  Fetch and cache text content from a URL.

  Downloads text from the given URL and saves it locally so that
  repeated calls use the cached version instead of re-downloading.
  Prints a success or failure message depending on the outcome.

  Parameters:
    raw_url (str): The URL of the raw text file to fetch.

  Returns:
    str: The full text content from the URL, or an empty string
         if the request fails.
  """
  import requests
  ...
  import requests
  from pathlib import Path
  import hashlib

  CACHE_DIR = Path("cs_110_content/text_cache")
  CACHE_DIR.mkdir(parents=True, exist_ok=True)

  def _url_to_filename(url):
    url_hash = hashlib.sha1(url.encode("utf-8")).hexdigest()[:12]
    return CACHE_DIR / f"{url_hash}.txt"

  cache_path = _url_to_filename(raw_url)

  SUCCESS_MSG = "✅ Text fetched."
  FAILURE_MSG = "❌ Failed to fetch text."
  try:
    if not cache_path.exists():
      response = requests.get(raw_url, timeout=10)
      response.raise_for_status()
      text_data = response.text
      cache_path.write_text(text_data, encoding="utf-8")
    print(SUCCESS_MSG)
    return cache_path.read_text(encoding="utf-8")

  except Exception as e:
    print(FAILURE_MSG)
    print(f"Error: {e}")
    return ""
```

### Save the text in a variable

```python
PRIDE_PREJUDICE_URL = "https://www.gutenberg.org/cache/epub/1342/pg1342.txt"

pride_prejudice_text = fetch_text(PRIDE_PREJUDICE_URL)
```

## Statistics about the data

```python
def print_text_stats(text):
  """
  Print basic statistics about a text string.

  Counts and displays the number of characters, lines, and words
  in the provided text.

  Parameters:
    text (str): The input text to analyze.

  Returns:
    None
  """

  num_chars = len(text)

  lines = text.splitlines()
  num_lines = len(lines)

  num_words = 0
  for line in lines:
    words_in_line = line.split()
    num_words_in_line = len(words_in_line)
    num_words += num_words_in_line

  print(f"Number of characters: {num_chars}")
  print(f"Number of lines: {num_lines}")
  print(f"Number of words: {num_words}")

print_text_stats(pride_prejudice_text)
```

```python
def get_word_counts(text):
  """
  Count the frequency of each word in a text string.

  Splits the text into lines and words, converts each word to
  lowercase, and returns a dictionary mapping each word to its
  frequency. No punctuation removal or stop word filtering is applied.

  Parameters:
    text (str): The input text to process.

  Returns:
    dict: A dictionary where keys are lowercase words (str) and
          values are their frequency counts (int).
  """
  word_counts = {}
  lines = text.splitlines()
  word_counts = {}
  lines = text.splitlines()
  for line in lines:
    words = line.split()
    for word in words:
      word = word.lower()
      if word in word_counts:
        word_counts[word] += 1
      else:
        word_counts[word] = 1
  return word_counts

word_counts = get_word_counts(pride_prejudice_text)
print(word_counts)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
def word_tokenization_normalization(text):
  """
  Tokenize and normalize text using spaCy NLP preprocessing.

  Converts text to lowercase, processes it with a spaCy language
  model, and filters out stop words, punctuation, numbers, and
  tokens with two or fewer characters. Returns the lemmatized
  form of each remaining token.

  Parameters:
    text (str): The raw input text to process.

  Returns:
    list: A list of lemmatized, normalized word strings.
  """
  text = text.lower() # lowercase
  doc = nlp(text)     # loading text into model

  words_normalized = []
  for word in doc:
    if word.text != '\n' \
    and not word.is_stop \
    and not word.is_punct \
    and not word.like_num \
    and len(word.text.strip()) > 2:
      word_lemmatized = str(word.lemma_)
      words_normalized.append(word_lemmatized)

  return words_normalized
```

```python
doc_tokenized = word_tokenization_normalization(pride_prejudice_text)
```

```python
# prompt: write a function that takes in a list of words and returns a word count dictionary

def word_count(word_list):
  """
  Count the frequency of each word in a list of words.

  Iterates through the provided list, converts each word to
  lowercase, and builds a dictionary mapping each unique word
  to the number of times it appears.

  Parameters:
    word_list (list): A list of word strings to count.

  Returns:
    dict: A dictionary where keys are lowercase words (str) and
          values are their frequency counts (int).
  """
  word_counts = {}
  for word in word_list:
    word = word.lower()
    if word in word_counts:
      word_counts[word] += 1
    else:
      word_counts[word] = 1
  return word_counts
```

```python
new_counts = word_count(doc_tokenized)
print(new_counts)
```

It removes stop words (common words like "the" and "and"), filters out punctuation and numbers, discards tokens of two characters or fewer, and applies lemmatization — reducing words like "says" and "said" to their base form "say." This significantly cleans the results.
The spaCy approach gives far more meaningful output for content analysis. Based on the top 15 words, we can clearly infer that Pride and Prejudice centers on characters named Elizabeth, Darcy, Bennet, Bingley, Jane, and Mrs., and that social interaction — through words like "say," "know," "think," "come," and "see" — is central to the story.
